```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
import cv2
import imutils
import matplotlib.pyplot as plt
from os import listdir
import time    

%matplotlib inline
```

```python
def augment_data(file_dir, n_generated_samples, save_to_dir):
    
    data_gen = ImageDataGenerator(rotation_range=10, 
                                  width_shift_range=0.1, 
                                  height_shift_range=0.1, 
                                  shear_range=0.1, 
                                  brightness_range=(0.3, 1.0),
                                  horizontal_flip=True, 
                                  vertical_flip=True, 
                                  fill_mode='nearest'
                                 )

    
    for filename in listdir(file_dir):
        print(filename)
        image = cv2.imread(file_dir + '/' + filename)
        image = image.reshape((1,)+image.shape)
        save_prefix = 'aug_' + filename[:-4]
        i=0
        for batch in data_gen.flow(x=image, batch_size=1, save_to_dir=save_to_dir, 
                                           save_prefix=save_prefix, save_format='jpg'):
            i += 1
            if i > n_generated_samples:
                break
```

```python
augmented_data_path = 'augmented_data/'

yes_path = 'data/yes'
no_path = 'data/no'

augment_data(file_dir=yes_path, n_generated_samples=6, save_to_dir=augmented_data_path+'aug_yes')
augment_data(file_dir=no_path, n_generated_samples=9, save_to_dir=augmented_data_path+'aug_no')
```

```python
aug_yes_path = 'augmented_data/aug_yes'
aug_no_path = 'augmented_data/aug_no'

m_aug_yes = len(listdir(aug_yes_path))
m_aug_no = len(listdir(aug_no_path))

m_aug = m_aug_yes + m_aug_no

print(m_aug)

print("% of tumor cases     = " , m_aug_yes*100/m_aug)
print("% of non-tumor cases = " , m_aug_no*100/m_aug)
```
